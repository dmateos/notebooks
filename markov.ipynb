```python
import random

def create_chain(str, batch=1):
    split = str.split(" ")
    counts = {n : {} for n in split}

    prev_word = ""
    for n in split:
        if prev_word != "":
            try:
                counts[prev_word][n] += 1
            except KeyError:
                counts[prev_word][n] = 1
        prev_word = n

    return counts

def _pick_word_from_list(l):
    flat_list = []
    for w, c in l.items():
        flat_list += [w]*c
    
    if flat_list == []:
        flat_list.append("an")

    choice = random.choice(flat_list)

    #print(flat_list)
    #print(choice)
    return choice

def generate_word(chain, length=100):
    start = list(chain)[0]
    for _ in range(0, length):
        word = _pick_word_from_list(chain[start])
        start = word
        yield word
        

COUNT=10000
training_data = ""
with open("hp.txt", "r") as f:
    training_data = f.readlines() 
training_data = " ".join(training_data)
training_data = training_data.replace("\n", "")
training_data = training_data.replace("  ", " ")
training_data = training_data.replace("-", " ")
training_data = training_data.replace("/", "")
training_data = training_data.lower()

chain = create_chain(training_data)
print(chain)
sentance = " ".join([n for n in generate_word(chain, COUNT)])
print(sentance)
```
